```python
# Convert 'DepDelay' to numeric and fill missing values
df['DepDelay'] = pd.to_numeric(df['DepDelay'], errors='coerce')
df['DepDelay'] = df['DepDelay'].fillna(df['DepDelay'].median())

# Convert 'Distance' to numeric and fill missing values
df['Distance'] = pd.to_numeric(df['Distance'], errors='coerce')
df['Distance'] = df['Distance'].fillna(df['Distance'].median())
```

```python
# Select relevant features for machine learning
df = df[['DepTime', 'ArrTime', 'ArrDelay', 'DepDelay', 'Distance', 'Delayed']]
```

```python
X = df.drop('Delayed', axis=1)  # Features
y = df['Delayed']              # Target
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Check the shapes of the splits
print(f"Training set: {X_train.shape}, Testing set: {X_test.shape}")
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, classification_report

# Initialize and train the model
model = LogisticRegression()
model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = model.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")
print(f"Precision: {precision:.2f}")

# Print a detailed classification report
print(classification_report(y_test, y_pred))
```
